화씨(F)로 된 새로운 열 avg_t_F를 만들어 보자.

𝐹=32+1.8×𝐶

```python
weather["avg_t_F"] = 32 + 1.8*weather["avg_t"]
weather
```

```python
#최고기온 평균(℃)의 최저기온 평균(℃)의 평균을 avg_t_minmix로 저장해보자
weather["avg_t_minmix"] = (weather["max_t"] + weather["min_t"])/2.0


weather["avg_t_minmix_2"] = weather[ ["max_t", "min_t"] ].sum(axis=1) /2.0
weather
```

```python
weather[ ["max_t", "min_t"] ].sum(axis=0) #열(0) 1:행 
```

```python
house = pd.read_csv("https://ilovedata.github.io/teaching/bigdata2/data/house_final.csv")
```

```python
#년단위로 그룹화하기
#groupby( by=...)는 데이터프레임을 물리적으로 그룹을 나누어 주는 것은 아니고 메소드에서 지정된 그룹을 데이터프레임에 적용하는 기능을 한다.
house_grp = house.groupby( by=["year"] )
```

```python
house_year = house_grp.sum().reset_index()
house_year
```

```python
#토탈 인덱스 만들기
house_year["total"] = house_year.sum(axis=1)
house_year
```

```python
#안전한 프로그래밍을 위하여 합을 구할 모든 열의 이름을 가진 리스트로 슬라이싱
house_year["total"] = house_year[ ["p1","p2","p3","p4","p5","p6","p7plus"] ].sum(axis=1)
house_year
```

```python
#1인 가구수에 대한 백분율(percentage)
house_year['p1_percent'] = house_year['p1'] / house_year['total'] *100
house_year
```

```python
house_year.plot.line(x='year', y='p1_percent')
```

```python
import numpy as np
import pandas as pd
```

```python
df3 = pd.DataFrame({
    "name" : ["철수", "영이"],
    "weight"  : [55, 44]
})
df3
```

```python
pd.merge(df1, df3, on="name") #식별자, 조건을 안 주면 공통된 식별자만 데이터 프레임으로 만든다
```

```python
pd.merge(df1, df3, on="name", how='left') #df1(왼쪽 데이터 자료)에 있는 식별자 다 뽑음
```

```python
df4 = pd.DataFrame({
    "name" : ["철수", "영이", "흥민"],
    "height"  : [167, 175, 183]
})
df4
```

```python
pd.merge(df1, df4, on="name", how='right') # 식별자는 오른쪽 데이터프레임만 있는 것으로
```

```python
pd.merge(df1, df4, on="name", how='inner') # 식별자는 두 데이터프레임에 공통인 것
```

```python
pd.merge(df1, df4, on="name", how='outer')  # 식별자는 두 데이터프레임의 모든 것
```

```python
#연도별 총인구수를 가진 열 pop_total 만들기
```

```python
print(pop_final) #sum 함수를 사용하면 문자열을 배제해줌
```

```python
pop_final["pop_total"] =  pop_final[["male", "female"]].sum(axis=1)
```

```python
pop_final
```

```python
all_final = pd.merge(house_final[['year', 'house_total']], pop_final[['year','pop_total']], on = 'year')
```
